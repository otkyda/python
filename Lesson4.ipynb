### Задание 2

### Создание массива t из 51 числа от 0 до 10 включительно с помощью функции linspace

```python
t = np.linspace(0, 10, 51)
t
```

### Создание массива f, содержащего косинусы элементов массива t

```python
f = np.cos(t)
f
```

### Построение линейной диаграммы, используя массив t для координат по горизонтали, а массив f - для координат по вертикали. Линия графика должна быть зеленого цвет, а ограничение графика по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5

```python
plt.plot(t, f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

### Задание 3

### Cоздание массива x из 51 числа от -3 до 3 включительно с помощью функции linspace библиотеки Numpy, а также массивов y1, y2, y3, y4 по формулам

```python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

 ### Создание объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(8, 6) #установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали
fig.subplots_adjust(wspace=0.3, hspace=0.3) #вертикальные и горизонтальные зазоры между графиками должны составлять 0.3
ax1, ax2, ax3, ax4 = ax.flatten() #преобразование массивов
ax1.plot(x, y1)
ax1.set_xlim([-5, 5]) #для графика в левом верхнем углу установите границы по оси x от -5 до 5
ax1.set_title('График y1') #в левом верхнем графике для координат по вертикали используйте y1
ax2.plot(x, y2)
ax2.set_title('График y2') #в правом верхнем - y2 
ax3.plot(x, y3)
ax3.set_title('График y3') #в левом нижнем - y3
ax4.plot(x, y4)
ax4.set_title('График y4') #в правом нижнем - y4
```
